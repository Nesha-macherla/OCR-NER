```python

!pip install pymupdf pytesseract spacy Pillow
!apt-get install -y poppler-utils
!apt-get install -y tesseract-ocr
!python -m spacy download en_core_web_sm

# Import necessary libraries
import fitz  # PyMuPDF
import pytesseract
from PIL import Image, ImageEnhance, ImageFilter
import io
import spacy
from google.colab import files

# Load spaCy's pre-trained NER model
nlp = spacy.load("en_core_web_sm")

def preprocess_image(image):
    # Convert to grayscale
    gray_image = image.convert('L')

    # Enhance the image (contrast, sharpness)
    enhancer = ImageEnhance.Contrast(gray_image)
    enhanced_image = enhancer.enhance(2.0)  # Adjust contrast level as needed
    enhancer = ImageEnhance.Sharpness(enhanced_image)
    enhanced_image = enhancer.enhance(2.0)  # Adjust sharpness level as needed

    # Apply a filter to reduce noise
    filtered_image = enhanced_image.filter(ImageFilter.MedianFilter())

    return filtered_image

def extract_text_from_scanned_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""

    for page_num in range(len(doc)):
        page = doc.load_page(page_num)
        pix = page.get_pixmap()
        img = Image.open(io.BytesIO(pix.tobytes()))

        # Preprocess the image for better OCR
        preprocessed_img = preprocess_image(img)

        # Extract text using pytesseract
        text += pytesseract.image_to_string(preprocessed_img)

    return text

def perform_ner(text):
    doc = nlp(text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    return entities

# Upload the PDF file
uploaded = files.upload()
pdf_path = next(iter(uploaded.keys()))

# Extract text from the scanned PDF
scanned_pdf_text = extract_text_from_scanned_pdf(pdf_path)
scanned_entities = perform_ner(scanned_pdf_text)

print("Extracted Text from Scanned PDF:\n", scanned_pdf_text)
print("\nNamed Entities:\n", scanned_entities)
```
